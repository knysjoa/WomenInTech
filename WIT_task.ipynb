```python
import re
#Import regular expressions operations library to make sure that each element provided by the user will have lower cases and no spaces

def isAnagram(s1, s2, s3):
    lower1 = re.sub("^\\s+|\\s+$|\\s+", "", s1.lower())
    lower2 = re.sub("^\\s+|\\s+$|\\s+", "", s2.lower())
    lower3 = re.sub("^\\s+|\\s+$|\\s+", "", s3.lower())
    print('s1 in lower case and no spaces-', lower1)
    print('s2 in lower case and no spaces-', lower2)
    print('s2 in lower case and no spaces-', lower3)
    s1 = sorted(lower1)
    s2 = sorted(lower2)
    s3 = sorted(lower3)
# Sorted() function returns a sorted list from the items in an iterable.
# Thanks to that we can easily compare all iterable elements defined by the user
    if s1 == s2 == s3:
       print("We have anagrams!")
    else:
       print("We do not have anagrams. ")
#We use == which means 'equal' and we are checking if all elements defined by the user are the same: a==b==c.
#If statement will be correct, user will see the notification "This is an anagram".
# In other situation user will see information that this is not an anagram.
def anagramCheck(s1,s2,s3):
    return sorted(s1) == sorted(s2) == sorted(s3)
#This soultion will return "False" if we do not have anagrams in our set or "True" if we have anagrams
#To create "more userfriendly" function, we can use if and print. Check def isAnagram.

print(anagramCheck('lala', 'LALA', 'laal'))
print(isAnagram('lala', 'LALA', 'laal'))
```
